_**Датасет с платформы Kaggle с данными о клиентах телекоммуникационной компании**_


**Признаки с их описанием:**

– ClientPeriod: количество месяцев использования услуг клиентом;

– MonthlySpending: ежемесячные расходы клиента на пользование услугами;

– TotalSpent: общие затраты клиента на пользование услугами компании;

– Sex («Male» / «Female»): пол клиента;

–  IsSeniorCitizen (1 / 0): является ли клиент иностранным гражданином;

– HasPartner («Yes» / «No»): есть ли у клиента партнер;

– HasChild («Yes» / «No»): есть ли у клиента ребенок;

– HasPhoneService («Yes» / «No»): подключена ли у клиента телефонная связь;

– HasMultiplePhoneNumbers («Yes» / «No»): имеет ли клиент несколько телефонных номеров;

– HasInternetService («Fiber optic» / «No» / «DSL»): имеет ли клиент доступ в интернет;

– HasOnlineSecurityService («Yes» / «No» / «No internet service»): подключена ли услуга онлайн службы безопасности;

– HasOnlineBackup («Yes» / «No» / «No internet service»): подключена ли услуга онлайн резервной копии;

– HasDeviceProtection («Yes» / «No» / «No internet service»): подключена ли услуга защиты устройства;

– HasTechSupportAccess («Yes» / «No» / «No internet service»): подключена ли услуга быстрого доступа к технической поддержке;

– HasOnlineTV («Yes» / «No» / «No internet service»): подключена ли услуга онлайн-телевидения;

– HasContractPhone («One year» / «Two year» / «Month-to-month»): тип контракта, заключенного с клиентом;

– IsBillingPaperless («Yes» / «No»): является ли выставление счетов для клиента безбумажным;

– PaymentMethod («Mailed check» / «Credit card (automatic)» / «Electronic check» / «Bank transfer (automatic)»): метод оплаты услуг;

– Churn (1 / 0): целевая переменная – откажется клиент от услуг компании или нет. (1 – отказался, 0 – нет)

Колонка TotalSpent не переводится в числовой тип данных, однако количество нулевых значений равно 0. Проверим TotalSpent на наличие числовых значений с помощью регулярного выражения и вычислим количество строк, где TotalSpent не содержит числовые данные.

## Создание новых признаков

**Клиент владеет несколькими пакетами услуг (1 - да, 0 -нет)**

```python

services = ["has_phone_service", "has_internet_service", "has_online_t_v", "has_multiple_phone_numbers"]

services_scored = df.loc[:, services].apply(lambda x: sum(1 if value == "Yes" or value == "Fiber optic" or value == "DSL" else 0 for value in x), axis=1)
df.loc[:, "has_multiple_services"] = (services_scored >= 2).astype('int64')
```

```python
df.head(1)
```

**Клиент подписан на дополнительные услуги (1 - да, 0 -нет)**

```python
subscriptions = ['has_online_security_service', 'has_online_backup', 'has_device_protection', 'has_tech_support_access', 'has_movie_subscription']

subs_scored =  df.loc[:, subscriptions].apply(lambda x: sum(1 if value == "Yes" or value == "Fiber optic" or value == "DSL" else 0 for value in x), axis=1)
df.loc[:, 'has_multiple_subscriptions'] = (subs_scored >= 2).astype('int64')
```

**Клиент состоит в отношениях и имеет ребенка (1 - да, 0 -нет)**

```python
df.loc[:, 'has_partner_and_child'] = (df.loc[:, 'has_partner'] == "Yes") & (df.loc[:, 'has_child'] == "Yes")
df.loc[:, 'has_partner_and_child'] = df.loc[:, 'has_partner_and_child'].astype('int64')
```

**Отношение ежемесячных платежей к общим тратам на услуги компании**

```python
df.loc[:, 'monthly_to_total_ratio'] = round(df.loc[:, 'monthly_spending'] / df.loc[:, 'total_spent'], 2) 
```

**Отношение периода к ежемесячным тратам**

```python
df.loc[:, 'period_to_monthly_ratio'] = round(df.loc[:, 'client_period'] / df.loc[:, 'monthly_spending'], 2)
```

### Кодирование категориальных данных

```python
def coding_columns(df, columns):
    for column in columns:
        df.loc[:, column] = df.loc[:, column].apply(lambda x: 1 if x=='Yes' else 0)
    return df
```

```python
columns_to_coding = ['has_partner', 'has_child', 'has_phone_service', 'has_multiple_phone_numbers',
                         'has_online_security_service', 'has_online_backup', 'has_device_protection', 'has_tech_support_access',
                         'has_online_t_v', 'has_movie_subscription', 'is_billing_paperless']
```

### Построение моделей

**Алгоритм Random Forest**

```python
#Импорт необходимых библиотек
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc
from sklearn.model_selection import GridSearchCV
```

```python
# Задаем seed для воспроизводимости результатов
import numpy as np
np.random.seed(123)
```

```python
# Разделение данных на обучающий и тестовый наборы
train_data, test_data = train_test_split(df_normalize, test_size=0.3)
```

```python
train_data.head(3)
```

```python
# Количество деревьев в лесу
ntree = 100
```

```python
# Количество переменных для случайного выбора в каждом разбиении
mtry = int(np.floor(np.sqrt(train_data.shape[1] - 1)))
```

```python
y = train_data['churn']
```

```python
# Создание  модели Random Forest
rf_model = RandomForestClassifier(n_estimators=ntree, max_features=mtry, random_state=123)
```

```python
# Выполнение кросс-валидации
cv_scores = cross_val_score(rf_model, train_data.drop('churn', axis=1), y, cv=5, scoring='accuracy')
print("Cross-validation scores:", cv_scores)
print("Mean accuracy:", cv_scores.mean())
```

```python
rf_model.fit(train_data.drop('churn', axis=1), y)
```

```python
# Прогнозирование на тестовых данных
predictions = rf_model.predict(test_data.drop('churn', axis=1))
test_data['predict'] = predictions
```

```python
test_data.head(3)
```

```python
# Оценка качества модели
accuracy = accuracy_score(test_data['churn'], predictions)
print("Accuracy:", round(accuracy, 3))
```

```python
# Построение Confusion Matrix
cm = confusion_matrix(test_data['churn'], predictions)
print("Confusion Matrix:\n", cm)
```

```python
# Точность (precision) и полнота (recall)
precision = precision_score(test_data['churn'], predictions)
recall = recall_score(test_data['churn'], predictions)
print("Precision:", round(precision, 3))
print("Recall:", round(recall, 3))
```

```python
# Расчет F1-score
f1 = f1_score(test_data['churn'], predictions)
print("F1-score:", round(f1, 3))
```

```python
# Построение ROC-кривой

test_data = test_data.drop('predict', axis=1)

rf_probs = rf_model.predict_proba(test_data.drop('churn', axis=1))[:, 1]

fpr, tpr, thresholds = roc_curve(test_data['churn'], rf_probs)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for RandomForest Model')
plt.legend(loc="lower right")
plt.show()
```

```python
# Просмотр наиболее значимых признаков
importances = rf_model.feature_importances_
feature_names = train_data.drop('churn', axis=1).columns
feature_importances = pd.DataFrame({'feature': feature_names, 'importance': importances})
print(feature_importances.sort_values(by='importance', ascending=False))
```

ROC-AUC измеряет качество бинарного классификатора по его способности разделять классы и оценивает общую способность модели отличать положительные и отрицательные классы.

ROC-кривая показывает отношение между долей верно классифицированных положительных случаев (true positive rate, TPR или recall) и долей ложноположительных случаев (false positive rate, FPR). ROC-AUC вычисляет площадь под ROC-кривой и дает общую оценку качества модели.

**Итог**: Наилучший показатель по критерию ROC-AUC имеет логистическая регрессия - 0,84.
